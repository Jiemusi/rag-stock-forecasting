```python
import os
import pandas as pd

DATA_DIR = "/Users/gaohan/rag-stock-forecasting/data/company_dataset"


symbols = sorted([f.replace(".csv", "") for f in os.listdir(DATA_DIR) if f.endswith(".csv")])
symbols
```

```python
for symbol in symbols:
    print("="*80)
    print(f"ðŸ“Œ {symbol} dataset info:\n")
    df = pd.read_csv(f"{DATA_DIR}/{symbol}.csv", parse_dates=["date"])
    print(df.info())
    print("\nMissing values per column:\n", df.isna().sum())
    print("\nDate range:", df['date'].min(), "-->", df['date'].max())
    print("="*80, "\n")
```
